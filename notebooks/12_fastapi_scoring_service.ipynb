# FastAPI Scoring Service

This is the first deployment step for the credit-risk project.

The goal is to turn the saved Random Forest champion pipeline into an API that can score new applicants outside Jupyter.

## 1. What This API Does

The API loads the saved model:

`models/random_forest_champion_pipeline.joblib`

Then it exposes endpoints for health checks, model summary, single applicant scoring, batch applicant scoring, and lightweight monitoring metrics.

## 2. API Endpoints

- `/health`: checks whether the API is running and whether the model file exists
- `/model/summary`: returns the champion model and decision policy
- `/predict`: scores one applicant
- `/batch_predict`: scores multiple applicants
- `/metrics`: returns simple request and prediction counters

## 3. Run the API Locally

From the project root, run:

```bash
uvicorn credit_risk_platform.api.main:app --reload
```

Then open:

```text
http://127.0.0.1:8000/docs
```

The `/docs` page gives you an interactive API testing page.

## 4. Environment Note

The saved model should be loaded with the same major dependency versions used when it was trained. The project dependency file pins scikit-learn to `>=1.7.2` because the champion model artifact was created with that version.

## 6. Expected API Response

The API returns three main outputs:

- `high_risk_probability`: model probability that the applicant is high risk
- `predicted_label`: low-risk or high-risk model class at the default 0.50 threshold
- `decision`: business decision band using approve, manual review, or reject

## 7. Why This Matters

This step moves the project from notebook analysis into a deployable scoring workflow.

Instead of manually running model code in Jupyter, a user or application can send applicant data to the API and receive a risk probability and decision recommendation.
